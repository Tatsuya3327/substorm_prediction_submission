# Import Libraries

In addition to well-known libraries, we use `dataHandler/AsimClassifier` from a research of aurora classification.

```python
from dataHandler import AsimClassifier
import numpy as np
import pandas as pd
import glob
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import sys
import os
import re
```

# Load and Classify Images 

```python
torch_data_location = "/Users/t.k/Desktop/Programming/Substorm_prediction/test_env/data/torch/"
clf = AsimClassifier(torch_path=torch_data_location)

results_dfs = []

for index, day_dir in enumerate(glob.glob('/Users/t.k/Desktop/Programming/Substorm_prediction/home/data/images/fsmi/*')):
    
    image_file_list = glob.glob(day_dir + '/*.png')
    results_dfs.append(clf.classify_images(image_file_list))
    
    print(f'complete lassifying directory {index+1}')
```

# Save Probabilities

```python
prob_dir = '/Users/t.k/Desktop/Programming/Substorm_prediction/data/probabilities/fsmi/'
columns = ['file', 'class_0', 'class_1', 'class_2', 'class_3', 'class_4', 'class_5']
col_name_mapping = {'class_0' : 'arc',
                    'class_1' : 'diffuse', 
                    'class_2' : 'discrete', 
                    'class_3' : 'cloudy', 
                    'class_4' : 'moon',
                    'class_5' : 'clear'}

probability_dfs = []

os.makedirs(prob_dir)

for index, result in enumerate(results_dfs):
    # extract probabilities and rename columns
    prob_df = result[columns]
    prob_df = prob_df.rename(columns=col_name_mapping)
    
    # sort by time
    prob_df['time'] = pd.to_datetime(prob_df['file'].apply(lambda x: int(re.findall(r'([\d]+)\.png$', x)[0])), format='%Y%m%d%H%M%S')
    sorted_prob = prob_df.sort_values('time').reset_index().set_index('time').drop('index', axis=1)
    
    # export as csv
    date_str = sorted_prob.index[0].strftime('%Y%m%d')
    sorted_prob.to_csv(prob_dir + 'probability_' + date_str + '.csv')
    
    # save for plotting
    probability_dfs.append(sorted_prob)
    
    print(f'complete saving DataFrame {index+1}')

sorted_prob_dfs = sorted(probability_dfs, key=lambda x: x.index[0])
```

```python
def plot_probability(prob_df, figsize=(17, 5), start_t=None, end_t=None):
    # select time
    if start_t is not None and end_t is not None:
        df = prob_df[(prob_df.index >= start_t) & (prob_df.index < end_t)]
    else:
        df = prob_df
        
    # extract date
    date = df.index[0].date()
    
    fig = plt.figure(figsize=figsize)
    ax = fig.gca()
    plt.plot(df.index, df['arc'], label='arc')
    plt.plot(df.index, df['diffuse'], label='diffuse')
    plt.plot(df.index, df['discrete'], label='discrete')
    plt.plot(df.index, df['cloudy'], label='cloudy')
    plt.plot(df.index, df['moon'], label='moon')
    plt.plot(df.index, df['clear'], label='clear')
    plt.ylim(ymin=0, ymax=1)
    plt.title(f'probability of {date}')
    plt.xlabel('Time')
    plt.ylabel('prob')
    ax.xaxis.set_major_locator(mdates.MinuteLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    plt.xticks(rotation=45)
    plt.grid()
    plt.legend()
    plt.show()
```

```python
# example
plot_probability(sorted_prob_dfs[0], figsize=(100, 5))
```
